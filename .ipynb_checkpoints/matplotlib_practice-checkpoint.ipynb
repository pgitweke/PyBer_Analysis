```python
# Import dependecies.
import matplotlib.pyplot as plt
```

```python
# Set the X-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the ploat with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")

# Create labels for the x and y axis.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

#Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# Create the plot.
#ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# Create labels for the x and y axis.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Add a y-axis limit.
ax.set_ylim(0,45)

# Add title.
ax.set_title("PyBer Fare by Month")

# Add grid lines.
ax.grid()

# Add legend
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create  the plot
plt.barh(x_axis, y_axis)
```

```python
# If you want the data on opposite axes, switch the arguments in the barh() function and run the cell again

# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

# Add the legend
plt.legend()

## Create a title
plt.title ("PyBer Fare by Month")

# Create labels for the x and y axis
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create  the plot with ax.plot()
fig, ax = plt.subplots ()
ax.barh(x_axis, y_axis)
```

```python
# If we want the data on opposite axes, we need to switch the arguments in the barh() function, as shown here
#Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh (y_axis, x_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")

# Create a title
ax.set_title("PyBer Fare by Month")

# Create labels for the x and y axis
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Invert the y-axis data.
plt.gca().invert_yaxis()

# Add the legend
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# When we use the plt.plot() function to create a scatter plot, 
# we need to add a lowercase "o" as a parameter inside the parentheses.
# This switches the plot from a line chart to a scatter plot.

plt.plot(x_axis, y_axis, "o")
```

```python
# Create Scatter plot
plt.plot(y_axis, x_axis, "o", c = "red", label = "Chicago")

# Create a title
plt.title("PyBer Fare by Month")

# Create labels for the x and y axis
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add the legend
plt.legend()

# Invert the y-axis data.
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis)

plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots ()
ax.scatter(y_axis, x_axis, linewidth=2, alpha = 0.8, label="Boston")

# Create title
ax.set_title("PyBer Fare by Month")

# Create labels for x and y
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Add a limit for x_axis.
ax.set_xlim(0.0,50)

# Invert the y-axis data.
plt.gca().invert_yaxis()

# Add legend
ax.legend()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie (y_axis, labels = x_axis)
plt.show
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values =(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct="%.1f%%")

plt.show()
```

```python
x = ["slateblue", "magenta", "lightblue", "grey", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax=plt.subplots()
fig.set_size_inches(8,8)
explode_highest_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
ax.pie(y_axis,
       labels=x_axis,
      autopct="%.1f%%",
      shadow=True,
       counterclock=False,
      startangle=100,
       colors= x,
       explode=explode_highest_values,
       rotatelabels=True,
      )


plt.show()
```
